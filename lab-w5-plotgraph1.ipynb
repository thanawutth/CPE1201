```python
names = ['Morning', 'Noon', 'Evening', 'Nigth']
values = [25, 31, 29,24]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Temperature Plotting')
plt.show()
```
